## Instantiating Lightning Module


```python
model_yolo = YOLOV3_PL(learning_rate=1e-3, num_epochs=40)

trainer = Trainer(max_epochs=40, precision='16', accelerator = "auto",devices="auto",
                  logger=CSVLogger(save_dir="/content/drive/MyDrive/Yolo/logs"),
                  check_val_every_n_epoch=8,
                  callbacks=[ LearningRateMonitor(logging_interval='step'),
                             TQDMProgressBar(refresh_rate=10),
                             ModelCheckpoint( dirpath='/content/drive/MyDrive/Yolo/checkpoints/',
                                             every_n_epochs=5,
                                             filename='{epoch}-{val_loss:.2f}' )])
```

```python
trainer.fit(model_yolo)
```

```python
torch.save(model_yolo.state_dict(), "/content/drive/MyDrive/Yolo/model.pth",_use_new_zipfile_serialization=False)
```

```python
metrics = pd.read_csv(f"{trainer.logger.log_dir}/metrics.csv")
del metrics["step"]
metrics.set_index("epoch", inplace=True)
display(metrics.dropna(axis=1, how="all").head())
sn.relplot(data=metrics, kind="line")
```

```python
model_yolo.load_state_dict(torch.load("model.pth", map_location=torch.device('cpu')), strict=False)
```
